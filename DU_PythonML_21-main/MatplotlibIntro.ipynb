```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
```

```python
data = {
    "names" : ["Akash","Aman","Ashok","Ashina","Deshna","Deepanshu","Rachit"],
    "marks" : [78,80,73,85,82,89,85]
}
```

```python
plt.figure(figsize=(10,6))
# plt.bar(data['names'], data['marks'])
plt.barh(data['names'], data['marks'], color='red')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(data['marks'], width=0.9)
plt.show()
```

```python
plt.figure(figsize=(10,6))
# plt.bar(data['names'], data['marks'])
plt.pie(labels=data['names'], x=data['marks'], 
        autopct='%1.2f%%', startangle=90)
plt.title('Marks of Students')
plt.legend()
plt.show()
```
