```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('summer.csv')
```

```python
df['Athlete'] = df['Athlete'].str.split(", ").str[::-1].str.join(" ").str.title()
```

```python
men_data = df[df['Gender'] == 'Men']
women_data = df[df['Gender'] == 'Women']
```

```python
pd.value_counts(men_data['Athlete']).head(10).reset_index()
```

```python
new_df = pd.value_counts(men_data['Athlete']).head(10).reset_index()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.bar(new_df['index'],new_df['Athlete'])
plt.xticks(rotation='vertical')
plt.title("Top Athlete of all time")
plt.xlabel("Athlete")
plt.ylabel("Medal Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation='vertical')
plt.title("Top Athlete of all time")
plt.xlabel("Athlete")
plt.ylabel("Medal Count")

men_data['Athlete'].value_counts().head(10).plot.bar(width=0.9, color=sns.color_palette('inferno'))

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation='vertical')
plt.title("Top Athlete of all time")
plt.xlabel("Athlete")
plt.ylabel("Medal Count")

axes = men_data['Athlete'].value_counts().head(10).plot.bar(width=0.9, color=sns.color_palette('inferno'))

for ax in axes.patches:
    axes.annotate(format(ax.get_height()), (ax.get_x()+0.3, ax.get_height() + 0.4))

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation='vertical')
sns.countplot(x='Sport', data=df)
plt.show()
```

```python
countries = df.groupby(['Country','Medal'])['Sport'].count().sort_values(ascending=False).reset_index()
```

```python
countries = countries.pivot('Country','Medal','Sport').fillna(0)
```

```python
top_10 = countries.sort_values(by='Gold', ascending=False).head(10)
```

```python
top_10
```

```python
top_10.plot.bar(width=0.8,color=['brown','gold','silver'])
figure = plt.gcf()
figure.set_size_inches(12,6)
plt.show()
```
