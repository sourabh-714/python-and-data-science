```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
import pandas as pd
```

```python
df = pd.read_csv('svm_data/data.csv', header=None)
```

```python
df.head()
```

```python
X = df.iloc[:,:2].values
y = df.iloc[:,-1].values
```

```python
plt.scatter(X[:,0], X[:,1],c=y)
plt.show()
```

```python
df.shape
```

```python
def draw_svm(X,y,penalty=1.0):
    plt.scatter(X[:,0], X[:,1],c=y)
    clf = SVC(kernel='linear', C=penalty)
    clf.fit(X,y)
    
    #gca - get current axes
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    x1 = np.linspace(xlim[0], xlim[1], 200)
    x2 = np.linspace(ylim[0], ylim[1], 200)
    
    xx,yy = np.meshgrid(x1,x2)
    x = np.vstack([xx.flatten(), yy.flatten()]).T
    z = clf.decision_function(x).reshape(xx.shape)
    
    ax.contour(xx,yy,z,colors='k',levels=[-1,0,1],alpha=0.5,
               linestyles=['--', '-', '--'])
    ax.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1],
              s = 100, linewidth = 1)
    plt.show()
    return clf
```

```python
draw_svm(X,y,0.0001)
```

```python
draw_svm(X,y,0.01)
```

```python
draw_svm(X,y,0.1)
```
