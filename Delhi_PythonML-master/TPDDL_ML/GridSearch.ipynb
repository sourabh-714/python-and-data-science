```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.5)
```

```python
pipe = make_pipeline(StandardScaler(), LogisticRegression(solver='liblinear'))
```

```python
penalty_options = ['l1','l2']
lambda_values = (0.001,0.01,0.1,0.5,1.0,5.0,10.0,20.0)
parameters = dict(logisticregression__penalty = penalty_options, logisticregression__C = lambda_values)
```

```python
grid_search = GridSearchCV(pipe, parameters, cv=5, scoring='accuracy',verbose=2)
```

```python
grid_search
```

```python
grid_search.fit(X, y)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```
