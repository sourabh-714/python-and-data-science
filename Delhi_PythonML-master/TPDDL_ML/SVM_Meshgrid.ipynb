```python
import numpy as np
from sklearn.datasets import make_moons
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
data = make_moons(n_samples=400, noise=0.4)
```

```python
X,y = data[0], data[1]
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
def plot_boundary(predict):
    x1 = np.arange(min(X[:,0]) - 1, max(X[:,0]) + 1, 0.01)
    x2 = np.arange(min(X[:,1]) - 1, max(X[:,1]) + 1, 0.01)
    
    xx,yy = np.meshgrid(x1,x2)
    z = np.array([xx.flatten(), yy.flatten()]).T
    pred = predict(z)
    pred = pred.reshape(xx.shape)
    plt.contourf(xx,yy,pred,cmap=ListedColormap(('cyan','red')))
    plt.scatter(X[:,0], X[:,1], c=y)
    plt.show()
```

```python
reg = LogisticRegression()
reg.fit(X,y)
```

```python
pred = reg.predict(X)
```

```python
accuracy_score(y, pred)
```

```python
plot_boundary(reg.predict)
```

```python
svm = SVC(kernel='rbf').fit(X,y)
```

```python
svm_pred = svm.predict(X)
```

```python
accuracy_score(y, svm_pred)
```

```python
plot_boundary(svm.predict)
```
